```python
def query(i, j, features):
    """
    i == year
    j == List of elements to subset
    features = list of variables to pull out and add up to generate a single column called total
    """
    df = censusdata.download('acs5', i, censusdata.censusgeo(j), features)
    data = pd.DataFrame(columns = ['total'])
    data['total'] = df.sum(axis = 1, skipna = True)
    return(data)

def by_area(i, area, features):
    """
    i == year
    area == 'county' or 'state'
    Features == list of columns codes
    """
    subsets = list([[('state', '39'),('county', '*')], [('state', '39')]])
    if (area == 'county'):
            j = subsets[0]
            data = query(i, j, features)
            return(data)
    if (area == 'state'):
            j = subsets[1]
            data1 = query(i, j, features)
            data = pd.DataFrame(data = {'total':[data1['total'][0]]}, 
                                index = ['39000'])
            return(data)
    else:
            return('chose between county or state')
```
